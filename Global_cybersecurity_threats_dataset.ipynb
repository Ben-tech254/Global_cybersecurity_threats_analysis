##**Definition of terms**
**1. Ransomware attack:**
A type of malicious cyberattack where hackers encrypt a victim's data or systems and demand a ransom payment in exchange for restoring access.

**2. Malware attack:** is any software intentionally designed to harm devices, steal data, or disrupt operations.

**3. Phishing attack:**  is a cyberattack where criminals impersonate legitimate entities (via email, SMS, calls, or fake websites) to trick victims into revealing sensitive information like passwords, credit card details, banking credentials.

**4. Man-in-the-middle attack:** this attack occurs when a hacker secretly intercepts and possibly alters communications between two parties who believe they are directly communicating with each other.

**5. SQL injection attack:** is a web security vulnerability that allows attackers to interfere with database queries by injecting malicious SQL code.


**6. DDoS attack:** is a malicious attempt to disrupt the normal traffic of a targeted server, service, or network by overwhelming the target or its surrounding infrastructure with a flood of Internet traffic.

**7. Zero-day vulnerability type :** is a previously unknown software/hardware flaw that hackers exploit before the vendor releases a software update to fix the specific vulnerability in the software/hardware.

**8. Unpatched software vulnerability type :** is a security flaw or weakness in software that has not been addressed through updates hence can be exploited by attackers to gain unauthorized access to systems.

**9. Social engineering vulnerability type :** is a weakness or flaw within a software application that can be exploited by malicious actors to compromise the confidentiality, integrity, or availability of a system.

```python
#Identifying the most common and least common attack types
print(data['Attack Type'].value_counts())
data['Attack Type'].value_counts().plot(kind='line', figsize  = (10,5), marker = 'o', markerfacecolor = 'red', markeredgecolor = 'red', markersize = 9)
plt.grid(alpha = 0.1)
plt.title('Most Common Attack Types')
plt.xlabel('Attack Type')
plt.ylabel('Frequency')
plt.show()
```

```python
#Attack types with the most losses incurred
losses_attack_types = data.groupby('Attack Type')['Financial Loss (in Million $)'].sum()/1000.0
print(losses_attack_types.sort_values(ascending = False))
losses_attack_types.plot(kind = 'line', figsize = (10,5),marker = 'o', markerfacecolor = 'red',
                                                                                              markeredgecolor = 'red', markersize = 9)
plt.grid(alpha = 0.1)
plt.title('Attack Types with the Most Losses')
plt.xlabel('Attack Type')
plt.ylabel('Financial Loss (in Billions $)')
plt.show()
```

* Avoid saving your passwords in password managers, moreso those that can enable access to confidential documents or to bank accounts.
* Deploy and properly configure a firewall to block unauthorized traffic, filter malicious connections and segment networks thereby preventing cyber intrusions hence guaranteeing some safety.
